```python
#Importa os pacotes
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

%matplotlib inline
```

**1.Carregar dataset com os dados tratados e mesclados**

```python
caminho_arquivo_tratado = "./v2/historico_precos_com_regioes.zip"
df = pd.read_csv(caminho_arquivo_tratado, parse_dates=["Data_Coleta"])
df.head()
```

**4.Analise Univariada**

**4.0 Funcoes**

**4.0.1 Tabela de Frequencia**

```python
def frequency_table(df, colname):
    grouped_series = df.groupby(colname).size()
    grouped_series.name = "Freq_Absoluta"
    df_frequency = grouped_series.to_frame()
    df_frequency["Freq_Relativa"] = df_frequency["Freq_Absoluta"] / len(df)
    df_frequency["Freq_Percentual"] = df_frequency["Freq_Relativa"] * 100.0
    return df_frequency
```

**4.0.2 Grafico de Frequencia Relativa**

```python
def plot_relative_frequency_chart(df):    
    freq_fig = plt.figure(figsize=(15,5))
    freq_chart = plt.bar(df.index, df["Freq_Relativa"], figure=freq_fig)
    plt.grid()
```

**4.1 Produto**

**4.1.1 Tabela de Frequencia**

```python
df_produto = frequency_table(df, "Produto")
df_produto
```

**4.1.2 Grafico de frequencia relativa**

```python
plot_relative_frequency_chart(df_produto)
```

**4.2 Ano Coleta**

**4.2.1 Tabela de Frequencia**

```python
df_ano_coleta = frequency_table(df, "Ano_Coleta")
df_ano_coleta
```

**4.2.2 Grafico de frequencia relativa**

```python
plot_relative_frequency_chart(df_ano_coleta)
```

**4.3 Regiao de Planejamento**

**4.3.1 Tabela de Frequencia**

```python
df_regiao = frequency_table(df, "Nome_Regiao")
df_regiao
```

**4.3.2 Grafico de Frequencia relativa**

```python
plot_relative_frequency_chart(df_regiao)
```

**4.4.2 Tabela de Frequencia**

**4.4.3 Grafico de Frequencia relativa**

```python
pesos_freq = np.zeros_like(df["Valor_Venda_Litro_Em_Reais"]) + 1. / df["Valor_Venda_Litro_Em_Reais"].size
hist_plot = df["Valor_Venda_Litro_Em_Reais"].hist(bins=10, label="Preços", figsize= (15, 5), weights=pesos_freq)
```

**4.4.4 ECDF = Função de distribuição cumulativa**

```python
def ecdf(data):
    #Quantidade de registros
    n = len(data)
    
    #dados do eixo x ordenados
    x = np.sort(data)
    
    #dados do eixo y 
    y = np.arange(1, n+1)/n
    
    return x,y
```

```python
#Calcula ECDF para os dados de preco
x_preco, y_preco = ecdf(df["Valor_Venda_Litro_Em_Reais"])
```

```python
#Configura o grafico 
plt.figure(figsize= (15, 5))
plt.xlabel("Preco")
plt.ylabel("ECDF")

plt.plot(x_preco, y_preco, marker = ".",  linestyle = "none")
plt.show()
```

**5.Análise Multivariada**

**5.0 Funcoes**

**5.0.1 Gerar Grafico de Data Coleta x Valor de Venda Por Produto**

```python
def plot_chart_data_valor_por_produto(df, produto):
    df_prod = df[df["Produto"] == produto]

    fig, ax = plt.subplots(figsize=(15,5))
    ax.set_title("Media de Precos - {} x Ano".format(produto))
    plt.plot(df_prod["Ano_Coleta"].unique(), df_prod.groupby(["Ano_Coleta"]).mean()["Valor_Venda_Litro_Em_Reais"], marker="o")
    plt.ylabel("Valor Venda (R$/l)")
    plt.xlabel("Ano")
    plt.grid()
    plt.show()
```

**5.0.2 Grafico de correlação entre Data Coleta x Valores de Venda de Produto Separado por Região**

```python
def plot_chart_data_valor_por_produto_e_regiao(df, produto):
    df_prod = df[df["Produto"] == produto]
    
    qtd_regioes = len(df_prod["Nome_Regiao"].unique())
    fig, ax = plt.subplots(figsize=(15,15))
    regioes = df_prod["Nome_Regiao"].unique()
    for idx in range(1, qtd_regioes + 1):
        regiao = regioes[idx - 1]
        df_regiao = df_prod[df_prod["Nome_Regiao"] == regiao]
        anos_regiao = df_regiao["Ano_Coleta"].unique()
        valores_regiao = df_regiao.groupby("Ano_Coleta").mean()["Valor_Venda_Litro_Em_Reais"]

        plt.plot(anos_regiao, valores_regiao, marker="o", label=regiao)
        plt.ylabel("Valor Venda (R$/l)") 
        plt.legend()      
    ax.grid()    
    plt.show()
```

**5.1.1 Correlação entre Data Coleta x Valores de Venda Por Produto (GASOLINA)**

```python
plot_chart_data_valor_por_produto(df, "GASOLINA")
```

**5.1.2 Correlação entre Data Coleta x Valores de Venda de GASOLINA Separado por Região**

```python
plot_chart_data_valor_por_produto_e_regiao(df, "GASOLINA")
```
